### <div style="background-color:#C8AD7F;font-family:newtimeroman;color:#333333;font-size:150%;text-align:center;border-radius:10px 10px;">Certains coll√®ges ont t'ils davantage de d√©crochages scolaire que d'autres ? </div>         

<b>R√©duction du nombre d'√©l√®ves entre la 6e et la 3e:</b>                                                                         
* Calcul du taux de d√©crochage : (Effectif¬†en¬†6e‚àíEffectif¬†en¬†3e)/Effectif¬†en¬†6e                                                 
* Tests statistiques : ANOVA pour comparer les taux de d√©crochage entre les coll√®ges.                                                        

```python
colonnes_a_selectionner = ['Rentr√©e scolaire','Num√©ro du coll√®ge',"Nombre d'√©l√®ves total", '6√®mes total','3√®mes total','6√®mes filles','6√®mes gar√ßons','3√®mes filles','3√®mes gar√ßons','R√©gion', 'Secteur']
df_selection = df[colonnes_a_selectionner]
# Calculer le taux de d√©crochage pour chaque coll√®ge pour les diff√©rentes ann√©es
df_selection['taux_decrochage'] = (df_selection['6√®mes total'] -df_selection['3√®mes total']) / df_selection['6√®mes total']*100
df_selection
```

```python
# La moyenne du taux de d√©crochage par coll√®ge
d√©crochage_coll√®ge= df_selection.groupby('Num√©ro du coll√®ge')['taux_decrochage'].mean()
d√©crochage_coll√®ge=d√©crochage_coll√®ge.reset_index()
d√©crochage_coll√®ge
```

```python
 #Filtrer et supprimer les lignes o√π 'taux_decrochage en %' est √©gal √† -inf
d√©crochage_coll√®ge = d√©crochage_coll√®ge[d√©crochage_coll√®ge['taux_decrochage'] != -np.inf]
d√©crochage_coll√®ge
```

```python
df_selec=d√©crochage_coll√®ge
df_selec.reset_index()
df_selec = df_selec[df_selec['taux_decrochage'] >= 0]
```

```python
# S√©lection al√©atoire de 10 coll√®ges
college_aleatoires = np.random.choice(df_selec.index, 10, replace=False)
# Extraction des taux de d√©crochage pour ces coll√®ges
decrochage_selection = df_selec.loc[college_aleatoires]
```

```python
decrochage_selection
```

```python
# Cr√©ation du graphique
plt.figure(figsize=(10, 6))
sns.barplot(x=decrochage_selection['Num√©ro du coll√®ge'], y=decrochage_selection['taux_decrochage'], palette='viridis')
plt.xlabel('Num√©ro du coll√®ge')
plt.ylabel('Taux de d√©crochage moyen')
plt.title('Taux de d√©crochage moyen pour 10 coll√®ges s√©lectionn√©s au hasard')
plt.xticks(rotation=45)
plt.show()
```

### Test Statistique - ANOVA

* <b>Test Statistique - ANOVA</b></br>
Utilisons une ANOVA pour comparer les taux de d√©crochage entre les coll√®ges. L'ANOVA nous permet de d√©terminer si les diff√©rences observ√©es entre les groupes sont statistiquement significatives.

* <b>Hypoth√®ses pour l'ANOVA</b></br>
<u>Hypoth√®se nulle (H0) :</u> Les taux de d√©crochage moyens sont √©gaux pour tous les coll√®ges.</br>
<u>Hypoth√®se alternative (H1) :</u> Au moins un coll√®ge a un taux de d√©crochage moyen diff√©rent.</br>



### V√©rification des Hypoth√®ses

```python
coll√®ge_dc = df_selection[df_selection['taux_decrochage'] != -np.inf]
coll√®ge_dc = df_selection[ df_selection['taux_decrochage'] >= 0]
```

```python
groupes = [coll√®ge_dc[coll√®ge_dc['Num√©ro du coll√®ge'] == nom]['taux_decrochage']
for nom in coll√®ge_dc['Num√©ro du coll√®ge'].unique()]
# Effectuer l'ANOVA
f_value, p_value = stats.f_oneway(*groupes)
print('R√©sultats de l\'ANOVA :')
print('F-value:', f_value)
print('p-value:', p_value)
```

<div style="background-color:#ebebe8;font-family:newtimeroman; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);border:2px solid #0A2342">
<p style="font-size:20px;  line-height: 1.5em; text-indent: 20px;">
    <b><u>R√©sultat</u></b></br>
    
<b>F-value :</b></br>
La F-value est une statistique qui mesure la variance entre les groupes par rapport √† la variance au sein des groupes. Une valeur F plus √©lev√©e indique une plus grande diff√©rence entre les moyennes des groupes.</br>

<b>p-value :</b></br>
Une p-value de 0.0 indique une probabilit√© extr√™mement faible que les diff√©rences observ√©es entre les taux de d√©crochage des coll√®ges soient dues au hasard. En pratique, une p-value aussi faible est arrondie √† z√©ro, ce qui signifie que nous avons des preuves extr√™mement solides contre l'hypoth√®se nulle.

</p></div>

<div style="background-color:#ebebe8;font-family:newtimeroman; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);border:2px solid #0A2342">
<p style="font-size:20px;  line-height: 1.5em; text-indent: 20px;">
‚ûú<b>Conclusion :</b></br>
<u>Rejet de l'hypoth√®se nulle (H0) :</u> √âtant donn√© que la p-value est bien inf√©rieure au seuil de signification commun√©ment utilis√© (0.05), nous rejetons l'hypoth√®se nulle.</br>
Diff√©rences entre les coll√®ges : Il y a des preuves statistiques significatives pour dire que les taux de d√©crochage moyens diff√®rent entre les coll√®ges. Cela signifie qu'au moins un coll√®ge a un taux de d√©crochage moyen diff√©rent des autres.

</p></div>

### <div style="background-color:#C8AD7F;font-family:newtimeroman;color:#333333;font-size:150%;text-align:center;border-radius:10px 10px;">Cette tendance a-t-elle un lien avec la taille du coll√®ge ? üéì</div>

‚û§ On va examiner la relation entre la taille des coll√®ges (nombre d'√©l√®ves) et le taux de d√©crochage. 

<b>Visualisation de la Relation:</b>
Avant d'appliquer des tests statistiques, visualisons la relation entre la taille du coll√®ge et le taux de d√©crochage.

```python
coll√®ge=df_selection.groupby('Num√©ro du coll√®ge').agg({"Nombre d'√©l√®ves total":'mean','taux_decrochage':'mean'})
coll√®ge
```

```python
#Filtre et suppression des lignes o√π 'taux_decrochage ' est √©gal √† -inf ou il a une valeur <0
coll√®ge = coll√®ge[coll√®ge['taux_decrochage'] != -np.inf]
coll√®ge = coll√®ge[coll√®ge['taux_decrochage'] >= 0]
coll√®ge
```

```python
# Scatter plot pour visualiser la relation entre la taille du coll√®ge et le taux de d√©crochage
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Nombre d'√©l√®ves total", y='taux_decrochage', data=coll√®ge)
plt.title('Relation entre la Taille du Coll√®ge et le Taux de D√©crochage')
plt.xlabel('Nombre Total d\'√âl√®ves')
plt.ylabel('Taux de D√©crochage')
plt.show()
```

#### Test de Corr√©lation de Pearson

Le test de corr√©lation de Pearson mesure la force et la direction de la relation lin√©aire entre deux variables continues.</br>
Hypoth√®ses du Test de Corr√©lation de Pearson :</br>
<u>Hypoth√®se nulle (H0) :</u> Il n'y a pas de corr√©lation lin√©aire entre le nombre d'√©l√®ves total et le taux de d√©crochage (r = 0).</br>
<u>Hypoth√®se alternative (H1) :</u> Il existe une corr√©lation lin√©aire entre le nombre d'√©l√®ves total et le taux de d√©crochage (r ‚â† 0).


```python
# Calcul de la corr√©lation
correlation, p_value = stats.pearsonr(coll√®ge["Nombre d'√©l√®ves total"],coll√®ge['taux_decrochage'])
print('Coefficient de Corr√©lation de Pearson : r=', correlation)
print('p-value:', p_value)
```

<div style="background-color:#ebebe8;font-family:newtimeroman; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);border:2px solid #0A2342">
<p style="font-size:20px;  line-height: 1.5em; text-indent: 20px;">
‚ûú<u><b>Conclusion:</b></u></br>
Puisque la p-value est tr√®s petite, nous rejetons l'hypoth√®se nulle (H0). Cela signifie que les donn√©es fournissent suffisamment de preuves pour conclure qu'il existe une corr√©lation statistiquement significative entre la taille du coll√®ge et le taux de d√©crochage.
</br>En r√©sum√©, nous rejetons l'hypoth√®se nulle et acceptons l'hypoth√®se alternative, ce qui indique qu'il y a bien une corr√©lation (bien que faible et n√©gative) entre la taille du coll√®ge et le taux de d√©crochage.
</p></div>

### Le test t de Student

le test t de student est utilis√© pour comparer les moyennes de deux √©chantillons ind√©pendants et d√©terminer s'il existe une diff√©rence statistiquement significative entre ces moyennes.
</br>Dans ce cas, vous comparez le taux de d√©crochage scolaire entre les gar√ßons et les filles.                                                         
<u>hypoth√®se nulle :</u> il n'y a pas de diff√©rence significative entre les taux de d√©crochage des gar√ßons et des filles.</br>
<u>hypoth√®se alternative :</u> il y a diff√©rence significative entre les taux de d√©crochage des gar√ßons et des filles

```python
# Test t de Student pour comparer les moyennes
t_statistic, p_value = stats.ttest_ind(df['taux_d√©crochage_gar√ßons'], df['taux_d√©crochage_filles'])

# Affichage des r√©sultats
print('Test t de Student :')
print('Statistic t :', t_statistic)
print('p-value :', p_value)
```

<b>Statistique t</b></br>
Le r√©sultat de la statistique t est 7.365138949626577. Ce nombre indique √† quel point les moyennes des √©chantillons des gar√ßons et des filles diff√®rent en termes d'unit√©s d'√©cart-type. Plus le nombre est grand en valeur absolue, plus la diff√©rence entre les groupes est significative.

<b>p-value</b></br>
Le p-value est 1.8388215796875454e-13, ce qui est extr√™mement petit. Cela signifie que la probabilit√© d'observer une diff√©rence aussi importante entre les moyennes des taux de d√©crochage des gar√ßons et des filles, simplement en raison du hasard (sous l'hypoth√®se nulle que les moyennes sont √©gales), est tr√®s faible.

##### a. Pr√©paration et visualisation des donn√©es

```python
region = df_selection[df_selection['taux_decrochage'] != -np.inf]
region = df_selection[ df_selection['taux_decrochage'] >= 0]
```

```python
somme_par_region = region.groupby('R√©gion')['taux_decrochage'].mean()
somme_par_region=somme_par_region.reset_index()
somme_par_region
```

```python
# Visualisation des taux de d√©crocheurs par r√©gion
sns.barplot(x='R√©gion', y='taux_decrochage', data=somme_par_region)
plt.xticks(rotation=90)
plt.show()
```

##### ANOVA                                                                                                                                                                                                                                     
1. Hypoth√®se nulle (H0): Les taux de d√©crochage moyens sont √©gaux pour toutes les r√©gions.                                       
2. Hypoth√®se alternative (H1): Au moins une r√©gion a un taux de d√©crochage moyen diff√©rent.

```python
groupes = [region[region['R√©gion'] == nom]['taux_decrochage'] for nom in region['R√©gion'].unique()]
# Effectuer l'ANOVA
f_value, p_value = stats.f_oneway(*groupes)
print('R√©sultats de l\'ANOVA :')
print('F-value:', f_value)
print('p-value:', p_value)
```

<b>P-value tr√®s faible :</b></br> La p-value tr√®s faible (6.4√ó10‚àí396.4√ó10 ‚àí39) indique qu'il y a des preuves extr√™mement solides pour rejeter l'hypoth√®se nulle. Cela signifie qu'il est tr√®s improbable que les diff√©rences observ√©es entre les taux de d√©crochage des r√©gions soient dues au hasard. En d'autres termes, au moins une r√©gion a un taux de d√©crochage moyen significativement diff√©rent des autres.

<b>a. visualisation les donn√©es:</b>

```python
Secteur0 = df_selection[df_selection['taux_decrochage'] != -np.inf]
Secteur0 = df_selection[ df_selection['taux_decrochage'] >= 0]
```

```python
secteur = Secteur0.groupby('Secteur')['taux_decrochage'].mean()
secteur=secteur.reset_index()
secteur
```

```python
sns.barplot(x='Secteur', y='taux_decrochage', data=secteur)
plt.xticks(rotation=90)
plt.show()
```

<b>b. Test t de Student pour comparer public vs priv√©</b></br>                                                                            
<u>Hypoth√®ses du Test :</u>                                                                                                           
* Hypoth√®se nulle (H0) : Les taux de d√©crochage moyens des coll√®ges publics et priv√©s sont √©gaux.                                 
* Hypoth√®se alternative (H1) : Les taux de d√©crochage moyens des coll√®ges publics et priv√©s sont diff√©rents.

```python
# Test t de Student pour comparer public vs priv√©
public =Secteur0[Secteur0['Secteur'] == 'PUBLIC']['taux_decrochage']
priv√© = Secteur0[Secteur0['Secteur'] == 'PRIVE']['taux_decrochage']
t_stat, p_val = stats.ttest_ind(public, priv√©)
print(f"t-statistic: {t_stat}, p-value: {p_val}")
```

<b>Valeur de la t-statistique :</b></br>
La valeur n√©gative de la t-statistique (-13.075) indique que le taux de d√©crochage moyen des coll√®ges publics est inf√©rieur √† celui des coll√®ges priv√©s. La magnitude √©lev√©e de cette valeur (en absolu) montre une grande diff√©rence entre les deux moyennes.

<b>p-value :</b></br>
La p-value extr√™mement faible (bien en dessous du seuil de signification de 0.05) indique que la diff√©rence observ√©e entre les taux de d√©crochage des coll√®ges publics et priv√©s est tr√®s peu probable due au hasard.
